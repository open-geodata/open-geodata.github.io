# Introdução

```python
import os
import shutil
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq
```

<br>

## Merge Data

```python
# Input path
input_path = os.path.join('..', 'data', 'individuais')

# Create List
list_xlsx = [f for f in os.listdir(input_path) if f.endswith('.xlsx')]
list_xlsx.sort()
#list_xlsx = list_xlsx[0:10]

# Loop
dfs = pd.DataFrame()
for f in list_xlsx:
    try:
        filename = os.path.basename(f)
        print(filename)
        df = pd.read_excel(os.path.join(input_path, f))
        df['cod_ponto_file'] = filename.split(' ')[0]
        df['data_inicio_file'] = filename.split(' ')[1]
        df['data_fim_file'] = (filename.split(' ')[2]).split('.')[0]
        
        dfs = dfs.append(df, ignore_index=True)
    except OSError as e:
        print(e)

# Results
df = dfs
df.head()
```

<br>

## Rename Columns

```python
# Results
df.head(2)
```

```python
# Dictonary to Rename
dict_columns = {
    # Identificação
    'Código Ponto'           : 'cod_ponto',
    'Cod_Interaguas'         : 'cod_interaguas',
    'Tipo Rede'              : 'tipo_rede',
    'Status Ponto'           : 'status_ponto',
    
    # Amostra
    'Data Coleta'            : 'data_coleta',
    'Hora Coleta'            : 'hora_coleta',
    'Parametro'              : 'parametro',
    'Sinal'                  : 'sinal',
    'Valor'                  : 'valor',
    'Unidade'                : 'unidade',
    'Tipo Parâmetro'         : 'tipo_parametro',
    
    # 
    'Sistema Hídrico'        : 'sistema_hidrico',
    'Tipo de Sistema Hídrico': 'sistema_hidrico_tipo',
    'CLASSE'                 : 'classe',
    
    # Datas
    'Inicio Operação'        : 'data_inicio_operacao',
    'Fim Operação'           : 'data_fim_operacao',
    'Período DE'             : 'data_inicio_query',
    'Período ATE'            : 'data_fim_query',
    
    # Localização
    'Latitude'               : 'latitude_dms',
    'Longitude'              : 'longitude_dms',
    'Altitude'               : 'altitude',
    'Localização'            : 'localizacao',
    'Captação'               : 'captacao',
    'Município'              : 'municipio',
    'UF'                     : 'estado',
    'UGRHI'                  : 'ugrhi',
    
    # File
    'cod_ponto_file'         : 'cod_ponto_file',
    'data_inicio_file'       : 'data_inicio_file',
    'data_fim_file'          : 'data_fim_file',
}

# Rename e Reorder
df = df.rename(columns=dict_columns)
df = df[[v for k, v in dict_columns.items()]]

# Columns
print(list(df.columns))

# Results
df.head(3)
```

<br>

## Ajusta Valores

```python
#set(df['valor'])
#= pd.to_numeric(df['valor'])
df['valor'] = df['valor'].str.replace(',', '.')
```

<br>

## Ajusta Datas

```python
# Ajuste Dates
df['data_coleta']          = pd.to_datetime(df['data_coleta'], format='%d/%m/%Y')
df['hora_coleta']          = pd.to_datetime(df['hora_coleta'], format='%H:%M').dt.time

df['data_inicio_operacao'] = pd.to_datetime(df['data_inicio_operacao'], format='%d/%m/%Y')
df['data_fim_operacao']    = pd.to_datetime(df['data_fim_operacao'], format='%d/%m/%Y')

df['data_inicio_query']    = pd.to_datetime(df['data_inicio_query'], format='%d/%m/%Y')
df['data_fim_query']       = pd.to_datetime(df['data_fim_query'], format='%d/%m/%Y')

df['data_inicio_file']     = pd.to_datetime(df['data_inicio_file'], format='%Y-%m-%d')
df['data_fim_file']        = pd.to_datetime(df['data_fim_file'], format='%Y-%m-%d')

# Results
df.head(2)
```

<br>

## Ajusta Coordenadas

```python
def dms2dd(coord_dms):
    coord_dms = coord_dms.strip()
    try:
        coord_deg = float(coord_dms.split(' ', maxsplit=2)[0])
        coord_min = float(coord_dms.split(' ', maxsplit=2)[1])
        coord_sec = float(coord_dms.split(' ', maxsplit=2)[2])
        coord_dd = coord_deg + (coord_min/60) + (coord_sec/3600)
        coord_dd = coord_dd * -1
    except:
        coord_dd = 0
    return coord_dd

dms2dd('22 13 52')
```

```python
# Coordenadas
df.loc[:, 'latitude_dd'] = df['latitude_dms'].astype(str).apply(lambda x: dms2dd(x))
df.loc[:, 'longitude_dd'] = df['longitude_dms'].astype(str).apply(lambda x: dms2dd(x))

# Seleciona os que tem coordenadas
df['tem_coordenadas'] = df.apply(lambda x: 0 if x['longitude_dd'] != 0 and x['latitude_dd'] != 0 else 1, axis = 1)

# Results
df.head(2)
```

<br>

## Compara nome dos arquivos

```python
# Compara nome do arquivo com dados da tabela
df['cod_ponto_test']   = df.apply(lambda x : 0 if x['cod_ponto'] == x['cod_ponto_file'] else 1, axis=1)
df['data_inicio_test'] = df.apply(lambda x : 0 if x['data_inicio_query'] == x['data_inicio_file'] else 1, axis=1)
df['data_fim_test']    = df.apply(lambda x : 0 if x['data_fim_query'] == x['data_fim_file'] else 1, axis=1)

# Tem Erro?
# 0 = não tem erro
# 1 = tem erro
erro_cod_ponto_test   = len(df[df['cod_ponto_test'] == 1])
erro_data_inicio_test = len(df[df['data_inicio_test'] == 1])
erro_data_fim_test    = len(df[df['data_fim_test'] == 1])

# Teste de erros
if erro_data_inicio_test+erro_data_fim_test+erro_cod_ponto_test == 0:
    print('Não tem erro')
    df.drop(['cod_ponto_file', 'data_inicio_file', 'data_fim_file'], axis=1, inplace=True)
    df.drop(['cod_ponto_test', 'data_inicio_test', 'data_fim_test'], axis=1, inplace=True)
else:
    # Results: avalia se tem algum erros nos arquivos individuais
    print(
        'Tem algum erro em:\0 = não tem erro\n1 = tem erro\n{} data início\n{} data fim\n{} código ponto'.format(
            erro_data_inicio_test,
            erro_data_fim_test,
            erro_cod_ponto_test,
        )
    )

# Results
df.reset_index(drop=True, inplace=True)
df.sort_values(by=['cod_ponto', 'data_coleta'], ascending=True, inplace=True)
df_amostras = df
df.head(2)
```
